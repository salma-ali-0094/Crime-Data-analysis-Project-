```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import plotly.express as px
import datetime as dt
%matplotlib inline 
```

```python
df=pd.read_csv('cleaned_Chicago_crime_2')
```

### creat date difference for progrees timeline

```python
df['Updated On'] = pd.to_datetime(df['Updated On'] , errors='coerce')
```

```python
df['Date'] = pd.to_datetime(df['Date'] , errors='coerce')
```

```python
df['Date Difference'] = (df['Updated On'] - df['Date']).dt.days
```

```python
df['Date Difference'] = df['Date Difference'] / 365
```

```python
df.head(20)
```

## extract time, day, month  

```python
df['Day'] = df['Date'].dt.day_name()
```

```python
df['Month'] = df['Date'].dt.month_name()
```

```python
df['time'] = df['Date'].dt.time
```

```python
df.head(5)
```

```python
df.drop('Date', axis=1, inplace=True)
```

```python
df.drop('Updated On', axis=1, inplace=True)
```

```python
df.drop('ID', axis=1, inplace=True)
```
